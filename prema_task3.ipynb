```python
label=preprocessing.LabelEncoder()
df['Species']=label.fit_transform(df['Species'])
df.head()
```

```python
sns.boxplot(data=df,orient="h",palette="Set2")
```

```python
x=df.drop('Species',axis=1).values
y=df['Species'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("Feature training data:", x_train.shape)
print("Target training data:",y_train.shape)
print("Feature test data:",x_test.shape)
print("target test data:",y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
d_train=dtree.fit(x_train,y_train)
```

```python
y_pred=dtree.predict(x_test)
```

```python
df1=pd.DataFrame({'Actual' :y_test,'Predicted' :y_pred})
df1.head()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
features=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
targets=['Iris-setosa','Iris-vesicolor','Iris-virginica']
```

```python
from sklearn.tree import plot_tree
from sklearn import tree
fig,ax=plt.subplots(nrows=1,ncols=1,figsize=(8,5),dpi=300)
tree.plot_tree(dtree,feature_names=features,class_names=targets,filled=True)
fig.savefig('DecisionTree.jpg')
```
